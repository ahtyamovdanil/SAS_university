#### __Логическая схема__  
Логическая схема содержит схему данных, которую необходимо реализовать в SAS, которая содержит таблицы измерений (4 шт.), таблицу фактов (1 шт.), отчет (1 шт.) 

```sas
options validvarname=v7; /* Controls the type of SAS variable names */

proc import datafile="source1.xlsx" 
        out=source_remont
        dbms=XLSX replace;
     sheet='Ремонт'; 
     getnames=yes;
     *mixed=yes;
        *stringdate=yes;
 
run;

data work.source_remont;
    set work.source_remont;
    rename VAR1=DATE VAR2=FULLNAME VAR3=VIN VAR4=OPERATION VAR5=HOURS VAR6=PRICE VAR7=RATIO;
    *informat VAR1 DDMMYY10.;
    *format VAR1 DDMMYYP10.;
    VAR6 = scan(VAR6, 1 , ' ', 'm');
run;

*set file2;

proc print 
    data=work.source_remont(obs=10);
    title "Ремонт";
run;
```

```sas
proc import datafile="source1.xlsx"
        out=source_mash
        dbms=xlsx
        replace;
     sheet='Mash';   
     getnames=yes;
run;

proc print 
    data=work.source_mash(obs=10);
    title "Mash";
run;
```
