#### __Логическая схема__  
Логическая схема содержит схему данных, которую необходимо реализовать в SAS, которая содержит таблицы измерений (4 шт.), таблицу фактов (1 шт.), отчет (1 шт.) 

```sas
options validvarname=v7; /* Controls the type of SAS variable names */

/* import sheet ремонт from source1.xlsx */
proc import datafile="source1.xlsx" 
        out=source_remont
        dbms=XLSX replace;
     sheet='Ремонт'; 
     getnames=yes;
     *mixed=yes;
        *stringdate=yes;
 
run;


data work.source_remont;
    set work.source_remont;
    rename VAR1=DATE VAR2=FULLNAME VAR3=VIN VAR4=OPERATION VAR5=HOURS VAR6=PRICE VAR7=RATIO;
    VAR1=VAR1-21916; /* date value correction */
    format VAR1 DDMMYY10.; /* formatting date output */
    VAR6 = scan(VAR6, 1 , ' ', 'm'); /* remove denomination from price */
run;

*set file2;

proc print 
    data=work.source_remont(obs=10);
    title "Ремонт";
run;
```

```sas
proc import datafile="source1.xlsx"
        out=source_mash
        dbms=xlsx
        replace;
     sheet='Mash';   
     getnames=yes;
run;

data work.source_mash;
    set work.source_mash;
    rename VAR1=DR_LICENSE VAR2=PASSPORT VAR3=FULLNAME VAR4=BRAND VAR5=MODEL VAR6=YEAR VAR7=VIN;
run;

proc print 
    data=work.source_mash(obs=10);
    title "Mash";
run;
```

### creating table for masters

```sas
proc sort 
        data=work.source_remont 
        out=work.masters nodupkey;
    by FULLNAME;
run;

data work.masters;
    retain ID FULLNAME;
    set work.masters (keep = FULLNAME);
    ID=_n_; 
run;
    
proc print 
        data=work.masters(obs=10);
        
    title "Masters";
run;
```
